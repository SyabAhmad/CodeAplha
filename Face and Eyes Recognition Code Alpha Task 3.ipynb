```
import cv2
import random
import matplotlib.pyplot as plt
# Load the face cascade classifier
frontalface = cv2.CascadeClassifier("data/haarcascade_frontalface_default.xml")
righteye = cv2.CascadeClassifier("data/haarcascade_righteye_2splits.xml")
smile =  cv2.CascadeClassifier("data/haarcascade_smile.xml")
eyeglasses =  cv2.CascadeClassifier("data/haarcascade_eye_tree_eyeglasses.xml")

def detectFace(img_path):
    # Read the image
    img = cv2.imread(img_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = frontalface.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

    if len(faces) == 0:
        return img  # No faces found, return the original image

    for (x, y, w, h) in faces:
        # Draw a rectangle around each detected face
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 1)

        roi_gray = gray[y:y + h, x:x + w]
        roi_color = img[y:y + h, x:x + w]

        righteyes = righteye.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in righteyes:
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 0, 255), 2)

        # Detect nose in the face ROI
        # smiles = smile.detectMultiScale(roi_gray)
        # for (nx, ny, nw, nh) in smiles:
        #     cv2.rectangle(roi_color, (nx, ny), (nx + nw, ny + nh), (255, 0, 0),1)

        # Detect ears in the face ROI
        eyeglasse = eyeglasses.detectMultiScale(roi_gray)
        for (ex, ey, ew, eh) in eyeglasse:
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 0, 255), 2)

    # Display the image with rectangles drawn around the faces
    # cv2.imshow("Face Detection", img)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
    cv2.waitKey(0)
    cv2.destroyAllWindows()


# Specify the path to the image you want to process
imglist = ['face 1.jpeg','face 2.jpeg','face 3.jpeg','face 4.jpeg','face 5.jpeg','face 6.jpeg','face 7.jpeg','face 8.jpeg']

img = random.choice(imglist)
# image_path = "face 2.jpeg"
detectFace(img)
```
